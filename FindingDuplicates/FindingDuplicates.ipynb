```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(file_path)

df.duplicated().sum()
```

```python
df_duplicados = df[df.duplicated(keep=False)] # Duplicados con sus primeras apariciones
print(df_duplicados.shape)
df_duplicados.head(10)
```

```python
df_duplicados[df_duplicados['ResponseId']==9]
```

```python
df_dup_subset = df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'], keep=False)]
df_dup_subset
```

```python
grupos = df_dup_subset.groupby(['MainBranch', 'Employment', 'RemoteWork'], dropna = False)
variabilidad_por_grupo = grupos.agg(lambda x: x.nunique())
variabilidad_por_grupo.head()
```

```python
conteo_paises = df_duplicados['Country'].value_counts().sort_values(ascending=False)
conteo_paises
```

```python
plt.figure(figsize=(12,6))
conteo_paises.plot(kind='bar')
plt.title("Número de duplicados por paises")
plt.xlabel("Country")
plt.ylabel("Cantidad de duplicados")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
paises_df = df['Country'].value_counts().sort_values(ascending=False)
paises_df
```

```python
plt.figure(figsize=(50, 30))
paises_df.plot(kind='bar')
plt.title('Países de origen de los encuestados')
plt.xlabel('País')
plt.ylabel('Número de encuestados originarios de cada País')
plt.tight_layout()
plt.show()
```
