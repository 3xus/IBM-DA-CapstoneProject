```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html'
data = requests.get(url).text
data
```

```python
soup = BeautifulSoup(data, "html.parser")
soup
```

```python
table = soup.find('table')
table
```

```python
for row in table.find_all('tr'):
    cols = row.find_all('td')
    programming_lenguage = cols[1].getText()
    salary = cols[3].getText()
    print("{}, {}".format(programming_lenguage, salary))
    
```

```python
language = []
annual_avg_salary = []

rows = table.find_all('tr')[1:]

for row in rows:
    cols = row.find_all('td')
    language.append(cols[1].get_text(strip=True))
    annual_avg_salary.append(cols[3].get_text(strip=True))
    
print(language)
print(annual_avg_salary)
    
```

```python
df = pd.DataFrame({'Language': language, 'Average Annual Salary': annual_avg_salary})
df
```

```python
df.to_csv('popular-languages.csv', index = False, encoding='utf-8')
```
