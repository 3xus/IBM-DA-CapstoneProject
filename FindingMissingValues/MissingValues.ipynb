```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df.isnull(),
            cbar = False,
            cmap = 'viridis',
            yticklabels = False)

plt.title("Mapa de calor de valor faltantes en el dataset")
plt.xlabel("Columnas")
plt.ylabel("Filas")
plt.show()
```

```python
missing_percent = df.isnull().mean() * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

plt.figure(figsize=(15, 20))
sns.barplot(y=missing_percent.index, x=missing_percent.values, palette="magma")
plt.title("Porcentaje de valores faltantes por columna")
plt.xlabel("% de valores faltantes")
plt.ylabel("Columnas")
plt.tight_layout()
plt.show()
```

```python
col_numericas = df.select_dtypes(include=['int64','float64'])
col_numericas
```

```python
missing_numeric = col_numericas.isnull().mean() * 100
missing_numeric = missing_numeric[missing_numeric > 0].sort_values(ascending=False)

# Visualizar
plt.figure(figsize=(10, 5))
sns.barplot(x=missing_numeric.index, y=missing_numeric.values, palette="viridis")
plt.title("Porcentaje de valores faltantes en columnas numéricas")
plt.ylabel("% de valores faltantes")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
